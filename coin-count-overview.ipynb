```python
from IPython.display import display

sql="""SELECT r.thread_id,t.date_created_4chan, t.topic,reply_count, top1, top2,top3,top4,top5,count1,count2,count3,count4,count5
FROM  ratings r, threads t where t.thread_id = r.thread_id AND r.has_results=1 AND reply_count >1 ORDER BY date_created_4chan DESC, count1 DESC
"""

#show complete dataframe
pd.options.display.max_columns = None
pd.set_option('display.max_rows', 500)


df =pd.read_sql(sql, conn)
display(df)
```
